```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load dataset
athletes= pd.read_csv("/home/vincey/Desktop/data anaylsis/athlete_events.csv")
athletes.head()
```

```python
# load second data set
regions= pd.read_csv("/home/vincey/Desktop/data anaylsis/noc_regions.csv")
regions.head()
```

```python
# join the dataframes
athletes_df=athletes.merge(regions, how='left', on = 'NOC')
athletes_df
```

```python
# columns names consistent
athletes_df.rename(columns={'region':'Region','notes':'Notes'}, inplace=True)
athletes_df.head()
```

```python
# Top countries participating
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# plot the above data in graph( top 10 countries that has participated in olympics)

plt.figure(figsize=(12,6))
# plt.xticks(rotation=20)
plt.title("Overall Participation by Country")
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set2')
```

```python
# age distribution of the participants

plt.figure(figsize=(12,6))
plt.title('Age distribution of the athletes')
plt.xlabel("Age")
plt.ylabel("Number of participants")
plt.hist(athletes_df.Age, bins=np.arange(10,80,2),color='orange',edgecolor='white')
```

```python
# male and female participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# pie lot for male and female athletes
plt.figure(figsize=(12,6))
plt.title("Gender Distribution")
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',startangle=150,shadow=True)
```

```python
# Total medals
athletes_df.Medal.value_counts()
```

```python
# Total number of female athletes in each olympics
female_participants = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season =='Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womanOlympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womanOlympics, palette='Spectral')
plt.title("Women Participation")
```

```python
# plot a line graph
part = womanOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title("Plot of Female Athletes over time")
```

```python
# gold medal atheletes
gold_medals =athletes_df[(athletes_df.Medal == 'Gold')]
gold_medals.head()
```

```python
# take only the values that are different from NaN
gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
```

```python
# gold beyond 60
gold_medals['ID'][gold_medals['Age'] > 60].count()
```

```python
# from which event did the gold medals come from
sporting_event = gold_medals['Sport'][gold_medals["Age"]> 60]
sporting_event
```

```python
# plot for sporting event

plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(x=sporting_event)
plt.title("Gold Medals for Athletes over 60 years")
```

```python
# gold medals from each country
gold_medals.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
# visualization of the table above
totalGoldMedals = gold_medals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x="index", y = "Medal", data=totalGoldMedals, height=5,kind='bar', palette='rocket')
g.despine(left = True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per Country")
```

```python
# rio olympics
max_year = athletes_df.Year.max()
print(max_year)
team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team
team_names.value_counts().head(10)
```

```python
# barplot of the above result
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel("Countrywise Medals for the year 2016")
```
