```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5,0.1,0.8,0.8])  # controls the left,bottom,width,height of the canvas(from 0 to 1)

axis.plot(x,y,'r')
```

```python
# we can draw subgraphs
fig, axes = plt.subplots(nrows = 1, ncols=2) # submap is of 1 row , 2 columns

for ax in axes:
    ax.plot(x,y,'r')
```

```python
# draw a picture or graph inside another graph

fig = plt.figure()

# controls the left,bottom,width,height of the canvas(from 0 to 1)

axes1 =fig.add_axes([0.1,0.1,0.8,0.8])  # big axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # small canvas

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
fig = plt.figure(figsize=(16,9),dpi=300)  # new graphic object

fig.add_subplot()

plt.plot(x,y,'r')
```

```python
fig, axes = plt.subplots()
axes.set_xlabel('x')
axes.set_ylabel('y')

axes.set_title("title of graph")
axes.plot(x,y,'r')
```

```python
ax.legend(['label1','label2'])

fig, axes = plt.subplots(dpi=200)
axes.set_xlabel('x')
axes.set_ylabel('y')

axes.set_title("title of graph")
axes.plot(x,x**2)
axes.plot(x,x**3)

axes.legend(["y=x**2", "y=x**3"], loc=2)
```

```python
# in matplotlib, you can set other propereties such as line color, transparency, and more 
fig, axes = plt.subplots(dpi=150)
axes.plot(x,x**2,color='red',alpha=.8)
axes.plot(x,x+2,color='green',alpha=.8)
axes.plot(x,x+3)
```

```python
fig, ax = plt.subplots(dpi=100)

# line width
ax.plot(x,x+1,color='blue',lw=0.25)
ax.plot(x,x+2,color='blue',lw=1)
ax.plot(x,x+3,color='blue',linewidth=1.5)
ax.plot(x,x+4,color='blue',linewidth=2)
```

```python
fig, ax = plt.subplots(dpi=100)

# line width
ax.plot(x,x+1,color='blue',lw=2, linestyle='-')
ax.plot(x,x+2,color='blue',lw=2,linestyle='-.')
ax.plot(x,x+3,color='blue',lw=2,linestyle=':')

line, = ax.plot(x, x+4, color='black', lw=1.50)
line.set_dashes([5,10,15,10])
```

```python
fig, ax = plt.subplots(dpi=100)

ax.plot(x, x+1, color='blue',marker='o')
ax.plot(x, x+2, color='blue',marker='+')
ax.plot(x, x+3, color='blue',marker='s')
ax.plot(x, x+4, color='blue',marker='1')

ax.plot(x, x+5, color='blue',marker='o', markersize=5)
ax.plot(x, x+6, color='blue',marker='o', markerfacecolor='red')
ax.plot(x, x+7, color='blue',marker='s', markersize=8, markerfacecolor='green')
```

```python
# set the canvas grid and axis range

fig, Axes= plt.subplots(1,2, figsize=(10,5))

Axes[0].plot(x,x**2,x,x**3, lw=2)
Axes[0].grid(True)

# limiting of data
Axes[1].plot(x,x**2,x,x**3, lw=2)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
Axes[1].grid(True)
```

# other 2D Graphics

```python
n = np.array([0,1,2,3,4,5])

fig, axes= plt.subplots(1,4, figsize=(16,5))
axes[0].set_title('scatter')
axes[0].scatter(x, x+0.25*np.random.randn(len(x)))

axes[1].set_title('step')
axes[1].step(n, n**2,lw=2)

axes[2].set_title('bar')
axes[2].bar(n, n**2,align='center', alpha=0.8)

axes[3].set_title('fill_between')
axes[3].fill_between(x, x**2, x**3, color='green', alpha=0.5)
```

```python
# Draw a radar chart

fig = plt.figure(figsize=(6,6))
ax= fig.add_axes([0.0,0.0,1,1], polar=True)
t=np.linspace(0,2 * np.pi, 100)
ax.plot(t,t, color='blue',lw=3)
```

```python
# Draw a histogram

n = np.random.randn(10000)
fig, axes = plt.subplots(1,2, figsize=(12,4))
axes[0].set_title('Histogram')
axes[0].hist(n)

axes[1].set_title('cummulative detailed histogram')
axes[1].hist(n, cumulative=True, bins=5)
```

```python
# Draw a contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0,3.0,delta)
y= np.arange(-2.0,2.0,delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1-Z2)*2

print(X)
print(Y)
```

```python
fig, ax = plt.subplots()
CS= ax.contour(X,Y,Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('contour map')
```

```python
# draw 3D surface image
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(14,6))

# specify the 3D graphics to draw, with projection = '3d'
ax = fig.add_subplot(1,2,1, projection='3d')
ax.plot_surface(X,Y,Z, rstride=4, cstride=4,linewidth=0)
```

# write a python program to create a pie chart of popularity of programming languages

```python
import matplotlib.pyplot as plt
# data to plot
languages = 'Python','Java','PHP',"Javascript",'C#','C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['#1f77b4','#ff7f0c','#2ca02c','#d62728','#9467bd','#8c564b']
```

```python
# explode 1st slice
explode = (0.1,0,0,0,0,0)
plt.pie(popularity, explode=explode, labels=languages, colors=colors,
       autopct='%1.1f%%', shadow=True)   # startangle=140 to change angle of the largest
```
