```python
# drop irrelevant columns
car = car.drop(['Engine Fuel Type','Market Category','Vehicle Style','Popularity','Number of Doors','Vehicle Size'],axis=1)
car.head(5)
```

```python
# renaming the column names
car = car.rename(columns={'Engine HP':'HP', 'Engine Cylinders':'Cylinders','highway MPG':'MPG-H','city mpg':'MPG-C','MSRP':'Price','Transmission Type':'Transmission','Driven_Wheels':'Drive Mode'})
car.head(5)
```

```python
# total number of rows and columns
car.shape
```

```python
# rows containing duplicated data
duplicate_rows_car = car[car.duplicated()]
print("Number of duplicate rows: ", duplicate_rows_car.shape )
```

```python
# count the rows before removing the data
car.count()
```

```python
# drop the duplicates
car = car.drop_duplicates()
# car.head()
car.count()
```

```python
# find the null values
print(car.isnull().sum())
```

```python
# drop the missing values(cleaning of data) or you can put the median
car = car.dropna()
car.count()
```

```python
print(car.isnull().sum())
```

```python
# finding the outliers
sns.boxplot(x=car['Price'])
```

```python
sns.boxplot(x=car['HP'])
```

```python
# plotting histogram for number of cars per brand
car.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('Number of cars by make')
plt.ylabel('Number of cars')
plt.xlabel('Make')
```

```python
# find the relationship between the varibales
plt.figure(figsize=(20,10))
c = car.corr()
sns.heatmap(c,cmap='BrBG', annot=True)
c
```
