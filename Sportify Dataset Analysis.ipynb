```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('/home/vincey/Desktop/data anaylsis/dataset.csv')
df_tracks.head()
```

```python
#check for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
sorted_df = df_tracks.sort_values('popularity', ascending =True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
most_popular =df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending =False)
most_popular[:10]
```

```python
df_tracks.set_index('track_genre', inplace=True)
# df_tracks.index= pd.datetime(df_tracks.index)  #if date is available
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[18]
```

```python
# converting duration from milisec to sec
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
corr_df =  df_tracks.drop(['key','mode','explicit'],axis =1).corr(method="pearson")
plt.figure(figsize=(18,8))
heatmap = sns.heatmap(corr_df,annot=True,fmt='.1g',center=0, cmap='inferno',linewidths=1,linecolor='blue')
heatmap.set_title('Correlation Heatmap Between Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y= 'loudness',x='energy',color='c').set(title='Loudness vs Energy Correlation')

# from the graph below there's a positive correlation btw loudness and energy, the higher the energy, the higher the loudness
```

```python
# another regression plot

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y= 'popularity',x='acousticness',color='b').set(title='Popularity vs Acousticness Correlation')
```

```python
# if the csv data has date

# df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
# df_tracks.dates = pd.to_datetime(df_tracks.dates)
# years= df_tracks.dates.dt.year

df_tracks['genres'] = df_tracks.index.get_level_values('track_genre')
# df_tracks.genres = pd.to_datetime(df_tracks.dates)
genres= df_tracks.genres
```

```python
# plotting songs counts vs year of song release
# sns.displot(years,discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year')

sns.displot(genres,discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per genre')
```

```python
total_dr = df_tracks.duration
fig_dims =(18,7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x = genres, y = total_dr, ax= ax, errwidth = False).set(title='Genres vs Duration')
plt.xticks(rotation=90)
```

```python
# line visualization of the above
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims =(10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=genres , y = total_dr, ax=ax).set(title='Genres vs Duration')
plt.xticks(rotation=90)
```

```python
df_feature = pd.read_csv('/home/vincey/Desktop/data anaylsis/SpotifyFeatures.csv')
df_feature.head()
```

```python
plt.title("Duration of the songs in different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data =df_feature)
plt.xlabel("duration in milli seconds")
plt.ylabel("Genres")

# observations
# the classical and world music have longer time while the children's music have shorter time
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous = df_feature.sort_values('popularity', ascending=False).head(10)  # ascending=false also means in descending order 
sns.barplot(y='genre',x='popularity', data = famous).set(title="Top 5 Genres by Popularity")
```
